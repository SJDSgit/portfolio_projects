```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data=pandas.read_csv("reviews.csv",parse_dates=['Timestamp']) # loading the data.
data.head()
```

## Ratings by Days

```python
data['Days']= data['Timestamp'].dt.date
day_avg=data.groupby(['Days']).mean()
day_avg.head()
```

```python
plt.figure(figsize=(25,5))
plt.plot(day_avg.index,day_avg['Rating'])
```

## Ratings by week

```python
data['Week']= data['Timestamp'].dt.strftime('%Y-%U')
week_avg=data.groupby(['Week']).mean()
week_avg.head()
```

```python
plt.figure(figsize=(25,5))
plt.plot(week_avg.index,week_avg['Rating'])
```

## Ratings by Month

```python
data['Month']= data['Timestamp'].dt.strftime('%Y-%m')
month_avg=data.groupby(['Month']).mean()
month_avg.head()
```

```python
plt.figure(figsize=(45,5))
plt.plot(month_avg.index,month_avg['Rating'])
```

## Average rating by month by course

```python
data['Month']=data['Timestamp'].dt.strftime('%Y-%m')
month_avg_crs=data.groupby(['Month','Course Name'])['Rating'].mean().unstack()
month_avg_crs.plot(figsize=(25,5))
```

## Which day are people the happiest?

```python
data['Weekday']=data['Timestamp'].dt.strftime('%A')
data['Daynumber']=data['Timestamp'].dt.strftime('%w')
weekday_avg=data.groupby(['Weekday','Daynumber']).mean()
weekday_avg=weekday_avg.sort_values('Daynumber')
plt.figure(figsize=(10,5))
plt.plot(weekday_avg.index.get_level_values(0),weekday_avg['Rating'])
weekday_avg
```

## No. of ratings by course

```python
pie_cht_ratings_course = data.groupby(['Course Name'])['Rating'].count()

plt.pie(pie_cht_ratings_course,labels=pie_cht_ratings_course.index,shadow=True,normalize=True)

pie_cht_ratings_course
```
